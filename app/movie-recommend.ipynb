```python
import pandas as pd
movie=pd.read_csv('movie-type.csv')
movie
```

```python
x=movie.drop(columns=['movie'])
x
```

```python
y=movie['movie']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x.values,y)
prediction=model.predict([[34,2500,0],[55,5000,1]])
prediction
```

```python
import joblib
joblib.dump(model,'movie-prediction.joblib')
```
